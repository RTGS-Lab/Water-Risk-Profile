```python
import xarray as xr
import pandas as pd
import os

# Define the base path where NetCDF files are stored
base_path = "./downloaded_files/"

# Define the range of years to process
years = range(1951, 2025)  # From 1951 to 2024

# Define Saint Paul, MN coordinates
lat_stp = 44.99345371654748
lon_stp = -93.17500486841008
# Initialize an empty list to store DataFrames
data_list = []

# Loop through each year and each month's file
for year in years:
    for month in range(1, 13):  # Loop through January (1) to December (12)
        file_path = os.path.join(base_path, f"{year}/ncdd-{year}{month:02d}-grd-scaled.nc")  # Adjust based on actual file naming

        # Check if the file exists before proceeding
        if os.path.exists(file_path):
            print(f"Processing: {file_path}")

            # Open the NetCDF dataset
            data = xr.open_dataset(file_path)

            # Find the closest available latitude and longitude for Saint Paul
            nearest_point = data.sel(lat=lat_stp, lon=lon_stp, method="nearest")

            # Convert to pandas DataFrame
            df = nearest_point.to_dataframe().reset_index()

            # Add Year and Month columns
            df["Year"] = year
            df["Month"] = month

            # Append to list
            data_list.append(df)

        else:
            print(f"File not found: {file_path}")

# Concatenate all DataFrames into one
if data_list:
    final_df = pd.concat(data_list, ignore_index=True)

    # Save to CSV
    csv_path = "Saintpaul_combined_data_NOAA.csv"
    final_df.to_csv(csv_path, index=False)
    print(f"Saint Paul data successfully saved to {csv_path}")
else:
    print("No data files were processed.")
```
